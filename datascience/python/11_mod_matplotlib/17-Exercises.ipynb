### Exercise 1. 

Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
# Exercise 1
from matplotlib import pyplot as plt
import pandas as pd

sales = pd.read_csv('data/company_sales_data.csv')
sales.head(10)
````

````python
plt.plot(sales['month_number'],sales['total_profit'])
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.show()
````

### Exercise 2. 

Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Sold units number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
# Exercise 2
plt.plot(sales['month_number'],sales['total_units'],'o:r', linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(['Total Units'],loc=4)
plt.show()
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
plt.plot(sales['month_number'],sales['facecream'], label="Face Cream")
plt.plot(sales['month_number'],sales['facewash'], label="Face Wash")
plt.plot(sales['month_number'],sales['toothpaste'], label="Toothpaste")
plt.plot(sales['month_number'],sales['bathingsoap'], label="Bathing Soap")
plt.plot(sales['month_number'],sales['shampoo'],label="Shampoo")
plt.plot(sales['month_number'],sales['moisturizer'],label="Moisturizer")
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend()
plt.show()
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.

````python
plt.scatter(sales['month_number'],sales['toothpaste'])
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Units Sold')
plt.grid(True)
plt.show()
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
plt.bar(sales['month_number']-.25,sales['facecream'], width=.25, label="Face Cream")
plt.bar(sales['month_number'],sales['facewash'], width=.25, label="Face Wash")
plt.legend()
plt.xticks(ticks=sales['month_number'])
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.show()
````

````python
fc_sum= sum(sales['facecream'])
fw_sum= sum(sales['facewash'])
toothp_sum= sum(sales['toothpaste'])
bs_sum= sum(sales['bathingsoap'])
sp_sum= sum(sales['shampoo'])
m_sum= sum(sales['moisturizer'])
salespie = [fc_sum, fw_sum, toothp_sum, bs_sum, sp_sum, m_sum]

labels = ['Face Cream','Face Wash','Toothpaste','Bathing Soap','Shampoo','Moisturizer']

fig = plt.figure()

plt.style.use('seaborn-muted')
plt.pie(salespie, labels=labels,autopct='%1.1f%%')
plt.title('Portion of Sales by Product')
plt.show()


fig.savefig('SalesPieChart.pdf')
````
